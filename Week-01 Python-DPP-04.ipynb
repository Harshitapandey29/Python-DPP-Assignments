Practical question :

Q2] Write a Python function compute_basic_stats that takes a list of numbers and returns a dictionary with the
minimum, maximum, and average.

```python
def compute_basic_stats(numbers):
    #Ensure the list is not empty to avoid errors:
    if not numbers:
        return {"min": None, "max": None, "average": None}
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers)/len(numbers)
    
    return {
        "min": minimum,
        "max": maximum,
        "average": average
    }

sample_data = [10, 20, 30, 40, 50]

#calling function:
stats = compute_basic_stats(sample_data)

#printing the output:
print("Computed basic Statistics:", stats)
```

robustness.

Practical Question:

Q8] Develop a Python script that uses the math_tools module to read numbers from a file and perform various
arithmetic operations. Include error handling for missing files or wrong file formats. 

```python
# math_tools.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero"
    return a / b

# Sample data

file_content = """
10 , 20
30 , 40
"""

# Function to read numbers from the sample data

def get_numbers(data):
    numbers = []
    lines = data.strip().split("\n")
    for line in lines:
        parts = line.split(",")
        try:
            num1 = float(parts[0])
            num2 = float(parts[1])
            numbers.append((num1, num2))
        except:
            print("Skipping invalid line:", line)
    return numbers

# Main program

def main():
    numbers = get_numbers(file_content)

    for num1, num2 in numbers:
        print(f"\nNumbers: {num1}, {num2}")
        print("Addition:", add(num1, num2))
        print("Subtraction:", subtract(num1, num2))
        print("Multiplication:", multiply(num1, num2))
        print("Division:", divide(num1, num2))

# Run
main()
```

Q9] Discuss a scenario where robust error handling is crucial in a Python application.

Practical Question:

Q10] Build a basic data processing application using the math_tools module that reads a dataset, applies a
function to clean the data, and calculates statistical measures. Handle exceptions like missing data and parse
errors. 

```python
# math_tools.py

def mean(numbers):
    return sum(numbers) / len(numbers)

def minimum(numbers):
    return min(numbers)

def maximum(numbers):
    return max(numbers)

# Sample dataset

dataset = [
    {"age": "25", "height": "175"},
    {"age": "30", "height": "180"},
    {"age": "",   "height": "165"},   # missing age
    {"age": "35", "height": "NaN"}    # invalid height
]

ages = []
heights = []

# Function to clean values

def clean(value):
    try:
        return float(value)
    except:
        return None

# Go through each row and collect valid numbers

for row in dataset:
    age = clean(row["age"])
    height = clean(row["height"])
    if age is not None:
        ages.append(age)
    if height is not None:
        heights.append(height)

print("Age Statistics:")
print("Mean Age:", mean(ages))
print("Min Age:", minimum(ages))
print("Max Age:", maximum(ages))

print("\nHeight Statistics:")
print("Mean Height:", mean(heights))
print("Min Height:", minimum(heights))
print("Max Height:", maximum(heights))
```
