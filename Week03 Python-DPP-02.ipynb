```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/hp/Downloads/Student Evaluation.csv")
df.head(10)
```

## Data Cleaning

```python
# Identify missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Replace missing "Score" values with average of that subject

mean_sub = df['Score'].mean()
mean_sub

df['Score'] = df.groupby('Subject')['Score'].fillna(mean_sub)
df
```

```python
# Replace missing "Hours_Studied" values with overall mean

df['Hours_Studied'].fillna(df['Hours_Studied'].mean(), inplace=True)
df
```

```python
# Drop rows where both Score and Hours_Studied are missing

df.dropna(subset=['Score', 'Hours_Studied'], how='all', inplace=True)
```

## Performance Categorization 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create Performance column

df['Performance'] = np.where(df['Score'] >= 80, 'High', 'Low')

# Count of each category

perf_count = df['Performance'].value_counts()
print("\nPerformance Category Count:\n", perf_count)

# Distribution of Performance by Subject
performance_distribution = df.groupby(['Subject', 'Performance']).size().unstack(fill_value=0)
print("\nPerformance Distribution by Subject:\n", performance_distribution)

# Optional Visualization
performance_distribution.plot(kind='bar', stacked=True)
plt.title('Performance Distribution by Subject')
plt.ylabel('Number of Students')
plt.show()
```

## Insights & Trends

```python
#Check correlation between Hours_Studied and Score
corr = df['Score'].corr(df['Hours_Studied'])
print("\nCorrelation between Study Hours and Score:", round(corr, 2))

# Does more study time mean higher scores?
sns.scatterplot(data=df, x='Hours_Studied', y='Score', hue='Performance')
plt.title("Study Hours vs Score")
plt.show()
```

```python
#One student who scored above 80 with fewer than 10 study hours
student_high_score_low_hours = df[(df['Score'] > 80) & (df['Hours_Studied'] < 10)]

# Display 
print("Student who scored above 80 with fewer than 10 study hours:")
print(student_high_score_low_hours.head(1))
```
