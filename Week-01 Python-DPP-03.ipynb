QUESTION 5] Find the union and intersection of these two sets: {1, 2, 8, 7} and {3, 6, 9,2,10}.

```python
# Ans]

set1 = {1,2,8,7}
set2 = {3,6,9,2,10}
set3 = set1.union(set2)
set4 = set1.intersection(set2)
print(set3)
print(set4)
```
