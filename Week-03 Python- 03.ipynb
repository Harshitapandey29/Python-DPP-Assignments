## Questions

Question 11] How do you create a bar chart showing total Total Amount for each Product Category?

```python
import matplotlib.pyplot as plt
```

```python
# Group by Product Category
category_sales = df.groupby('Product Category')['Total Amount'].sum()

# Plot bar chart
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 12] How do you create a line plot of monthly total Total Amount using the Date column (resampled by month)?

```python
# Convert Date to datetime (in-memory)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Group by month and sum Total Amount
monthly_sales = df.resample('M', on='Date')['Total Amount'].sum()

# Plot line chart
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Question 13] How do you create a histogram showing the distribution of Total Amount?

```python
# Plot histogram for Total Amount
plt.figure(figsize=(8,5))
plt.hist(df['Total Amount'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Total Amount per Transaction')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
