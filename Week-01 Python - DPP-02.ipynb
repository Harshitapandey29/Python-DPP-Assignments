Question 6] Write a Python program that asks the user for a number and checks if it is even or odd.
Test Input: Number = 7
Expected Output: 7 is an odd number.

```python
#Ans] 

a = int(input("Enter a number:"))

if a%2 == 0:
    print(f"{a} is an even number")
else:
    print(f"{a} is an odd number")
```

Question 7] Write a Python program that takes a number from the user and prints its square and square root.

```python
#Ans] 

num = int(input("Enter a number:"))
square = num ** 2
square_root = num ** 0.5
print("Square:", square)
print("Square Root:", square_root)
```

Question 8] Write a Python program that takes a string from the user and prints:
The string in uppercase
The string in reverse order
The length of the string
Test Input: String = python

```python
#Ans] 

text = input("Enter a string: ")

print("Uppercase:", text.upper())
print("Reversed:", text[::-1])
print("Length:", len(text))
```
