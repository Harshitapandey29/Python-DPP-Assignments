```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Q1. Load Titanic dataset using seaborn :

```python
df = sns.load_dataset("titanic")
```

## Displaying top 5 rows :

```python
df.head()
```

## How many rows and columns does the dataset contain?

```python
df.shape
```

## Q2. Bar Chart :

```python
# Plot bar chart showing the number of passengers in each class ( ).

df['pclass'].value_counts().plot(kind='bar', color = "orange")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Passengers in Each Class")
plt.show()
```

## Q3. Boxplot :

```python
# Plot a boxplot of across different passenger classes (pclass):

sns.boxplot(x='pclass', y='age', data=df)
plt.title("Age Distribution Across Passenger Classes")
plt.show()
```

## Q4. Heatmap :

```python
# heatmap showing correlations among numerical columns

corr = df[['age', 'fare', 'sibsp', 'parch']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Q5. Pie Chart :

```python
# pie chart showing the proportion of passengers who survived vs. did not survive

df['survived'].value_counts().plot(kind='pie', labels=['Not Survived', 'Survived'])
plt.title("Survival Proportion")
plt.ylabel("")
plt.show()
```

```python
# Add percentages to each slice.

df['survived'].value_counts().plot(kind='pie', autopct = "%1.1f%%", labels=['Not Survived', 'Survived'])
plt.title("Survival Proportion")
plt.ylabel("")
plt.show()
```

## Q6. Scatterplot :

```python
# Create a scatterplot of age vs. fare, colored by sex.

sns.scatterplot(x='age', y='fare', hue='sex', data=df)
plt.title("Age vs Fare by Gender")
plt.show()
```

## Q7. Histogram :

```python
# Plot a histogram of age.

sns.histplot(df['age'], bins=10)
plt.xlabel("Age")
plt.title("Age Distribution")
plt.show()
```

## Q8. Line Chart :

```python
# Group the dataset by age and calculate the average fare paid at each age.

avg_fare_age = df.groupby('age')['fare'].mean()
avg_fare_age
```

```python
# Plot a line chart showing the trend of fare vs. age

plt.plot(avg_fare_age.index, avg_fare_age.values)
plt.xlabel("Age")
plt.ylabel("Average Fare")
plt.title("Average Fare Paid at Each Age")
plt.show()
```
