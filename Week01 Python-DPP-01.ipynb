Q5] Take a string input from the user and check if it’s a palindrome (same forwards and backwards).

```python
text = input("Enter a string:")
clear_str = text.replace(" ", "").lower()
if clear_str == clear_str[::-1]:
    print("It is a Palindrome.")
else:
    print("Not a Palindrome.")
```

Q12] Write a program to count the frequency of each word in a given string using a dictionary.

```python
text = input("Enter a string:")
words = text.split()
frequency = {}
for i in words:
    i = i.lower()
    frequency[i] = frequency.get(i,0)+1
    
for i, count in frequency.items():
    print(f"{i}: {count}")
```

Q13] Create a nested dictionary to store student names and their respective scores in three subjects. Print the
student with the highest total score.

```python
students = {
    "Ram": {"Math": 85, "Science": 92, "English": 78},
    "Rimmy": {"Math": 75, "Science": 88, "English": 90},
    "Siya": {"Math": 95, "Science": 85, "English": 80}
}

top_student = None
highest_score = 0

for name, subjects in students.items():
    total = sum(subjects.values())
    print(f"{name}'s total score = {total}")
    
    if total > highest_score:
        highest_score = total
        top_student = name
        
print(f"\nThe student with the highest total score is {top_student} with {highest_score} marks.")
```

Q14] Given a dictionary of employee names and salaries, update the salary of a specific employee based on user
input.

```python
employees = {
    "Ram": 50000,
    "Rimmy": 60000,
    "Rahul": 55000,
    "Kriti": 70000
}

print("Current employee salaries:", employees)

name = input("Enter the employee name to update salary: ")

if name in employees:
    new_salary = int(input(f"Enter new salary for {name}: "))
    employees[name] = new_salary
    print(f"Updated salary of {name}: {employees[name]}")
else:
    print("Employee not found!")

print("Updated employee salaries:", employees)
```

Q15] Set Operations:Write a program to find the union, intersection, and difference of two sets:
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

union_set = set1.union(set2)  #union
intersection_set = set1.intersection(set2)  #intersection
difference_set1 = set1 - set2  # Elements in set1 but not in set2
difference_set2 = set2 - set1  # Elements in set2 but not in set1

print("Set 1:", set1)
print("Set 2:", set2)
print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (set1 - set2):", difference_set1)
print("Difference (set2 - set1):", difference_set2)
```
