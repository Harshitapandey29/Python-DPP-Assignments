## Question 1

A company introduces a new customer feedback process. The mean satisfaction score before implementing
the process was 72. After implementing the process, a random sample of 30 customers gave a mean
satisfaction score of 78, with a standard deviation of 10. At a 5% significance level, test if the new process has
improved customer satisfaction.

## Question 2

A school claims that their students' average math score is higher than the national average of 75. A random
sample of 50 students scored an average of 77 with a standard deviation of 8. Perform a one-tailed test at a
0.01 significance level to verify the school’s claim.

## Question 3] 

A retail store wants to check if their new sales strategy increases the average sales from ₹10,000 per day. After
implementing the strategy for 15 days, the sales data showed an average of ₹11,200 per day with a standard
deviation of ₹1,500. Conduct a one-sample t-test at a 5% significance level.

## Question 4] 

A factory produces light bulbs with an advertised lifespan of 1200 hours. A sample of 40 bulbs is tested, and the
mean lifespan is found to be 1180 hours with a standard deviation of 50 hours. Test if the advertised lifespan is
accurate at a 0.05 significance level.

## Question 5] 

A researcher wants to test if the average height of adult males in a city differs from the national average of 5.8
feet. They collected data from 25 individuals and found an average height of 5.7 feet with a standard deviation
of 0.3 feet. Perform a two-tailed test at a 1% significance level.
